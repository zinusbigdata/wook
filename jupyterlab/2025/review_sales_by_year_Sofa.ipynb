## Heatmap for Category by Years

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("review-vc_sales_by_collection.csv")
```

```python
df
```

```python
df['collection'] = df['collection'].replace({'Aidan Sling Chair. 2PK':'Aidan Sling Chair, 2PK'})
```

```python
df['year'] = df['yr_month'].astype(str).str[:4]
```

## Sofa

```python
sofa_df = df[
    (df['financial_category']=='Sofa') &
    (df['year'].isin(['2022','2023','2024']))
].copy()
```

```python
sofa_df = sofa_df[sofa_df['collection'] != '__TOTAL__']
```

```python
sofa_df = sofa_df.dropna(subset=['written_avg_rating'])
```

```python
sofa_df[sofa_df['collection'].str.contains('Aidan Sling Chair',na=False)]
```

```python
sofa_df
```

```python
top_cols = (
    sofa_df.groupby('collection')['sales_amount']
    .sum()
    .sort_values(ascending=False)
    .head(20)
    .index
)
```

```python
sorted_cols = sorted(top_cols)
```

```python
sorted_cols
```

```python
sofa_cols = sofa_df[sofa_df['collection'].isin(sorted_cols)].copy()
```

```python
sofa_cols['rating_bin'] = sofa_cols['written_avg_rating'].round(1)
```

```python
sofa_cols
```

```python
for year in ['2022','2023','2024']:
    pivot = (
        sofa_cols[sofa_cols['year']==year]
        .groupby(['collection','rating_bin'])['sales_amount']
        .sum()
        .unstack(fill_value=0)
        #.sort_index()
    )
    pivot = pivot.loc[pivot.sum(axis=1).sort_values(ascending=False).index]

    mask = pivot == 0

    plt.figure(figsize=(6,6))
    sns.heatmap(pivot, cmap='YlOrRd', mask=mask, linewidths=0.5, linecolor='gray', cbar_kws={'label': 'Sales'})
                                                                                  
    plt.title(f'Sofa ({year})', fontsize=14)
    plt.xlabel('Review Rating')
    plt.ylabel('Collection')
    plt.tight_layout()
    plt.show()
    
```
