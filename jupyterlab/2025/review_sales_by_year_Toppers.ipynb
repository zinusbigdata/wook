```python
topper_df['rating_bin'] = topper_df['written_avg_rating'].round(1)
```

```python
for year in ['2022','2023','2024']:
    
    pivot = (
        topper_df[topper_df['year']==year]
        .groupby(['collection','rating_bin'])['sales_amount']
        .sum()
        .unstack(fill_value=0)
       # .sort_index()
    )
    
    pivot = pivot.loc[pivot.sum(axis=1).sort_values(ascending=False).index]

    plt.figure(figsize=(14,5))
    sns.heatmap(pivot, cmap='YlOrRd', linewidths=0.5, linecolor='gray', cbar_kws={'label': 'Sales'})
                                                                                  
    plt.title(f'Toppers - Review and Sales Heatmap ({year})', fontsize=14)
    plt.xlabel('Review Rating')
    plt.ylabel('Collection')
    plt.tight_layout()
    plt.show()
    
```
