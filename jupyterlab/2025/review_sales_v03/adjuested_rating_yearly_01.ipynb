# 년도별 Adjusted Rating 계산

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('collection_yearly_out1_rating_sales_2_4.csv')
```

```python
df_2022 = df[['collection','2022_rating','2022_sales']].copy()
```

```python
df_2022['2022_weight'] = df_2022['2022_rating'] * df_2022['2022_sales']
```

```python
df_2022
```

```python
total_weight_2022 = df_2022['2022_weight'].sum()
df_2022['2022_contribution'] = df_2022['2022_weight'] / total_weight_2022*100
#df_2022['2022_contribution'] = df_2022['2022_contribution'].round(4)
```

```python
df_2022['2022_contribution'].sum()
```

```python
df_2022 = df_2022.sort_values(by='2022_contribution', ascending=False)
df_2022['2022_acc_contribution'] = df_2022['2022_contribution'].cumsum()
```

```python
df_2022
```

```python
top_80 = df_2022[df_2022['2022_acc_contribution'] <= 80]
```
