```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('review-vc_sales-by_collection_category_yearly_20250428_v2.csv')
```

```python
# collection이 '__TOTAL__'이고 collection2도 '__TOTAL__'인 경우 제외
df = df[~((df['collection'] == '__TOTAL__') & (df['collection2'] == '__TOTAL__'))]
```

```python
df['collection'].nunique()
```

```python
# is_single_collection이 True인 경우만 필터링
filtered_df = df[df['is_single_collection'] == True]
single_col_df = filtered_df[['yr', 'financial_category', 'collection', 'written_avg_rating', 'sales_amount']]
```

```python
single_col_df['collection'].nunique()
```

```python
wonder=single_col_df[single_col_df['collection'].str.contains('WonderBox', na=False)] #['collection'].nunique()
wonder['collection'].nunique()
```

```python
# is_single_collection이 False이고, collection2가 '__TOTAL__'인 경우만 추출
filtered_df = df[
    (df['is_single_collection'] == False) & (df['collection2'] == '__TOTAL__')
]

wonderbox_col_df = filtered_df[['yr', 'financial_category', 'collection', 'written_avg_rating', 'sales_amount']].copy()

# collection 이름에 "(WonderBox)" 추가
wonderbox_col_df['collection'] = wonderbox_col_df['collection'] + ' (+WonderBox)'
```

```python
wonderbox_col_df['collection'].nunique()
```

```python
merged_df = pd.concat([single_col_df, wonderbox_col_df], ignore_index=True)
```
